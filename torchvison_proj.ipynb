```python
from pathlib import Path
import torch
import matplotlib.pyplot as plt
import requests
plt.rcParams["savefig.bbox"] = 'tight'

from torchvision.transforms import v2
from torchvision.io import decode_image

torch.manual_seed(1)


# 1. Define the remote URL and local filename
url = "https://raw.githubusercontent.com/pytorch/vision/main/gallery/assets/astronaut.jpg"
filename = "astronaut.jpg"

# 2. Download the image if it doesn't exist
if not Path(filename).exists():
    print(f"⬇️ Downloading {filename}...")
    response = requests.get(url)
    with open(filename, "wb") as f:
        f.write(response.content)

# 3. FIX: Point to the file in the CURRENT directory, not '../assets'
# Old Code: str(Path('../assets') / 'astronaut.jpg')
# New Code: filename
img = decode_image(filename)

print(f"✅ Success! Image Loaded.")
print(f"Type: {type(img)}, Shape: {img.shape}, Dtype: {img.dtype}")

# 4. Verify with your plot function
from helpers import plot
plot(img)
```
